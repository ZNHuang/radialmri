```python
import numpy as np
from scipy.io import loadmat
import matplotlib.pyplot as plt

kspace = loadmat("../sample/kspace/BC33_cartesian_origsmap_None_4.88e-09.mat")
print(kspace.keys())
print(kspace['simulated_kspace_cartesian'].shape)
```

```python
t = 10
c = 6
plt.imshow(np.fft.fftshift(np.sqrt(kspace['simulated_kspace_cartesian'][t, c, 0]**2 + kspace['simulated_kspace_cartesian'][t, c, 1]**2)))
plt.show()
```

```python
t = 10
c = 10
kspace_temp = (kspace['simulated_kspace_cartesian'][t, c, 0] + 1j*kspace['simulated_kspace_cartesian'][t, c, 1])
plt.imshow(np.abs((np.fft.ifft2(kspace_temp))))
plt.show()
```
